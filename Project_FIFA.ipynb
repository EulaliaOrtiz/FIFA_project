```python
def stHead(df):   #def standard_header(head):    #head = df.columns
    new_header = []
    for h in df.columns:
        #print(h)
        x = h.lower().replace(' ','_')
        #print(h)
        new_header.append(x)
    df.columns = new_header
    display(df)
```
